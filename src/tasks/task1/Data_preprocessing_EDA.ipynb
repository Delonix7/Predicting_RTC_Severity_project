```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pyarrow.parquet as pq
#authenticate colabs to allow access to Google Drive
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
```

```python
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)

def get_data_parquet(url, file_name):
  data_url_list = url.split('/')
  downloaded = drive.CreateFile({'id':data_url_list[5]})
  downloaded.GetContentFile(file_name)
  df = pq.read_table(file_name).to_pandas()
  return df

url_accident = 'https://drive.google.com/file/d/17Ezmmd5ANytFy3QlS8z6ZFVV79itPZqc/view?usp=sharing'
url_vehicle = 'https://drive.google.com/file/d/17QNs99r4RUIG9P5apEMzHVxSxjDVHMg_/view?usp=sharing'
file_name_accident = 'accident_data.parquet'
file_name_vehicle = 'vehicle_data.parquet'
```

```python
accident_data = get_data_parquet(url_accident, file_name_accident)
accident_data.shape
```

```python
vehicle_data = get_data_parquet(url_vehicle, file_name_vehicle)
vehicle_data.shape
```
